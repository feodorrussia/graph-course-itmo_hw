1. **Группы**

Множество целых чисел $\mathbb{Z}$ с операцией сложения образует **абелеву группу**.  
- **Замкнутость**: сумма любых двух целых чисел - целое число (из этого же множества).  
- **Ассоциативность**: $(a+b)+c = a+(b+c)$;
- **Нейтральный элемент**: $0$ (так как $a+0 = a$);
- **Обратный элемент**: для каждого $a$ существует $-a$, такой что $a + (-a) = 0$ (поэтому абелева);
- **Коммутативность**: $a+b = b+a$, поэтому группа абелева.

2. **Кольца и поля**

Множество квадратных матриц $2 \times 2$ с действительными элементами с операциями сложения и умножения матриц образует **кольцо**, но **не поле**.

Кольцо, так как:
- Сложение образует абелеву группу;
- Умножение ассоциативно и дистрибутивно относительно сложения.

Не поле, так как:
- Умножение не коммутативно в общем случае;
- И не каждая ненулевая матрица обратима (например, матрица с нулевым определителем не имеет обратной).

3. **Полиномы как кольцо**

Множество многочленов степени не выше $2$ с действительными коэффициентами **не является кольцом** относительно обычных сложения и умножения.
- При сложении замкнутость сохраняется (степень суммы не превышает максимум степеней);
- При умножении произведение двух многочленов степени $2$ может иметь степень $4$, что выходит за пределы множества. Нарушается условие замкнутости умножения (результат операции находится вне исходного множества), поэтому это не кольцо.

4. **Умножение матриц**

Даны матрицы:
$$
A = \begin{pmatrix}1 & 2 \\ 3 & 4\end{pmatrix},\quad 
B = \begin{pmatrix}0 & -1 \\ 2 & 5\end{pmatrix},\quad 
C = \begin{pmatrix}2 & 0 \\ 1 & 3\end{pmatrix}.
$$

Вычислим $(A \cdot B) \cdot C$:
$$
A \cdot B = \begin{pmatrix}1\cdot0+2\cdot2 & 1\cdot(-1)+2\cdot5 \\ 3\cdot0+4\cdot2 & 3\cdot(-1)+4\cdot5\end{pmatrix} = \begin{pmatrix}4 & 9 \\ 8 & 17\end{pmatrix},
$$
$$
(A \cdot B) \cdot C = \begin{pmatrix}4 & 9 \\ 8 & 17\end{pmatrix} \begin{pmatrix}2 & 0 \\ 1 & 3\end{pmatrix} = \begin{pmatrix}4\cdot2+9\cdot1 & 4\cdot0+9\cdot3 \\ 8\cdot2+17\cdot1 & 8\cdot0+17\cdot3\end{pmatrix} = \begin{pmatrix}17 & 27 \\ 33 & 51\end{pmatrix}.
$$

Вычислим $A \cdot (B \cdot C)$:
$$
B \cdot C = \begin{pmatrix}0\cdot2+(-1)\cdot1 & 0\cdot0+(-1)\cdot3 \\ 2\cdot2+5\cdot1 & 2\cdot0+5\cdot3\end{pmatrix} = \begin{pmatrix}-1 & -3 \\ 9 & 15\end{pmatrix},
$$
$$
A \cdot (B \cdot C) = \begin{pmatrix}1 & 2 \\ 3 & 4\end{pmatrix} \begin{pmatrix}-1 & -3 \\ 9 & 15\end{pmatrix} = \begin{pmatrix}1\cdot(-1)+2\cdot9 & 1\cdot(-3)+2\cdot15 \\ 3\cdot(-1)+4\cdot9 & 3\cdot(-3)+4\cdot15\end{pmatrix} = \begin{pmatrix}17 & 27 \\ 33 & 51\end{pmatrix}.
$$

Результаты совпадают, что подтверждает **ассоциативность умножения матриц**.

5. **Умножение Кронекера**

Для матриц
$$
A = \begin{pmatrix}1 & 2 \\ 3 & 4\end{pmatrix},\quad 
B = \begin{pmatrix}0 & -1 \\ 2 & 5\end{pmatrix}
$$
произведение Кронекера $A \otimes B$ вычисляется как блочная матрица:
$$
A \otimes B = \begin{pmatrix}
1\cdot B & 2\cdot B \\
3\cdot B & 4\cdot B
\end{pmatrix} = \begin{pmatrix}
\begin{pmatrix}0 & -1 \\ 2 & 5\end{pmatrix} & \begin{pmatrix}0 & -2 \\ 4 & 10\end{pmatrix} \\
\begin{pmatrix}0 & -3 \\ 6 & 15\end{pmatrix} & \begin{pmatrix}0 & -4 \\ 8 & 20\end{pmatrix}
\end{pmatrix} = \begin{pmatrix}
0 & -1 & 0 & -2 \\
2 & 5 & 4 & 10 \\
0 & -3 & 0 & -4 \\
6 & 15 & 8 & 20
\end{pmatrix}.
$$

6. **Алгоритм Штрассена**

Умножим матрицы $F = \begin{pmatrix}1 & 2 \\ 3 & 4\end{pmatrix}$ и $G = \begin{pmatrix}5 & 6 \\ 7 & 8\end{pmatrix}$ методом Штрассена.

Обозначим:
$$
F = \begin{pmatrix}a & b \\ c & d\end{pmatrix},\quad a=1,\; b=2,\; c=3,\; d=4,
$$
$$
G = \begin{pmatrix}e & f \\ g & h\end{pmatrix},\quad e=5,\; f=6,\; g=7,\; h=8.
$$

Вычисляем семь вспомогательных произведений:
$$
\begin{aligned}
M_1 &= (a+d)(e+h) = (1+4)(5+8) = 5 \cdot 13 = 65,\\
M_2 &= (c+d)e = (3+4)\cdot5 = 7 \cdot 5 = 35,\\
M_3 &= a(f-h) = 1 \cdot (6-8) = -2,\\
M_4 &= d(g-e) = 4 \cdot (7-5) = 8,\\
M_5 &= (a+b)h = (1+2)\cdot8 = 3 \cdot 8 = 24,\\
M_6 &= (c-a)(e+f) = (3-1)(5+6) = 2 \cdot 11 = 22,\\
M_7 &= (b-d)(g+h) = (2-4)(7+8) = (-2) \cdot 15 = -30.
\end{aligned}
$$

Теперь элементы произведения $C = F \cdot G$:
$$
\begin{aligned}
c_{11} &= M_1 + M_4 - M_5 + M_7 = 65 + 8 - 24 - 30 = 19,\\
c_{12} &= M_3 + M_5 = -2 + 24 = 22,\\
c_{21} &= M_2 + M_4 = 35 + 8 = 43,\\
c_{22} &= M_1 - M_2 + M_3 + M_6 = 65 - 35 - 2 + 22 = 50.
\end{aligned}
$$

Итак, $F \cdot G = \begin{pmatrix}19 & 22 \\ 43 & 50\end{pmatrix}$, что совпадает с результатом обычного умножения.

7. **Сравнение сложности для матриц $4 \times 4$**

**Стандартный метод** (прямое умножение):
- Число умножений: $n^3 = 4^3 = 64$.
- Число сложений: $n^2(n-1) = 16 \cdot 3 = 48$.
- Всего операций: $64 + 48 = 112$.

**Метод Штрассена** (одноуровневое разбиение на блоки $2 \times 2$):
- Каждое из 7 умножений блоков выполняется стандартным способом (для $2 \times 2$ требуется 8 умножений и 4 сложения на блок).  
  Итого умножений: $7 \times 8 = 56$.  
  Сложений от умножений блоков: $7 \times 4 = 28$.
- Дополнительные сложения/вычитания блоков в формулах Штрассена: 18 операций над блоками, каждая операция над блоками $2 \times 2$ требует 4 поэлементных сложений.  
  Итого: $18 \times 4 = 72$ сложения.
- Всего сложений: $28 + 72 = 100$.  
- Всего операций: $56 + 100 = 156$.

Для $n=4$ стандартный метод эффективнее (112 < 156). Однако для больших $n$ метод Штрассена выигрывает за счёт асимптотики $O(n^{\log_2 7}) \approx O(n^{2.807})$.

8. **Быстрое возведение в степень**

Алгоритм быстрого возведения квадратной матрицы в степень (exponentiation by squaring) основан на следующих свойствах:
- **Ассоциативность умножения матриц**: позволяет перегруппировывать множители.
- **Существование единичной матрицы** $I$: $A^0 = I$.
- Свойство степеней: $A^{2k} = (A^k)^2$ и $A^{2k+1} = A \cdot (A^{2k})$.

Благодаря этому можно сократить число умножений: вместо последовательного умножения (требующего $n-1$ умножений для степени $n$) используется рекурсивное деление показателя пополам, что даёт $O(\log n)$ умножений. Например, для $A^4$:
- Наивно: $((A \cdot A) \cdot A) \cdot A$ - 3 умножения.
- Быстро: $T = A \cdot A$ (1 умножение), затем $R = T \cdot T$ (ещё 1) - всего 2 умножения.
